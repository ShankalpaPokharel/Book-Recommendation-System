Filtering out less active users and less popular books enhances the accuracy and relevance of book recommendations by focusing on meaningful interactions. 

* Calculates the counts of unique users and ISBNs in the DataFrame df_ratings.
* Filters the df_ratings DataFrame to exclude rows where the user has rated less than 200 times and the ISBN has been rated less than 100 times.

```python
isbn_count = ratings['ISBN'].value_counts()
user_count = ratings['User-ID'].value_counts()
```

```python
ratings = ratings[~ratings['User-ID'].isin(user_count[user_count < 200].index) & ~ratings['ISBN'].isin(isbn_count[isbn_count < 100].index)]
```

```python
ratings.rename(columns={"ISBN":'isbn',"User-ID":"user","Book-Rating":'brating'}, inplace=True)
```

```python
ratings
```

A pivot table is a data analysis technique that arranges and summarizes data in a structured format, allowing you to group and aggregate information to gain insights and patterns from your data. It's particularly useful for transforming data into a more manageable and informative layout.

```python
df_table = ratings.pivot_table(index='isbn', columns='user', values='brating').fillna(0)
df_table
```

```python
df_table.index = df_table.join(books.set_index('ISBN'))['Book-Title']
df_table
```

```python
from sklearn.neighbors import NearestNeighbors
def get_recommends(book = ""):
    recommended_books = []  # Initialize an empty list to store recommended books and distances
    
    # Creating a NearestNeighbors model with cosine distance metric
    nbrs = NearestNeighbors(n_neighbors=6, metric="cosine").fit(df_table.values)
    
    # Finding nearest neighbors for the specified book
    distances, indices = nbrs.kneighbors([df_table.loc[book].values], n_neighbors=6)
    
    for i in range(1,6):  # Iterate over the 5 nearest neighbors (excluding the book itself)
        # Appending recommended book index and its distance to the list
        recommended_books.append([df_table.index[indices[0][-i]], distances[0][-i]])
    
    # Returning the list of recommendations, including the original book
    return [book, recommended_books]
```

```python
get_recommends("Politically Correct Bedtime Stories: Modern Tales for Our Life and Times")
```
